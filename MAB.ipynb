```python
# Imports
import numpy as np
```

```python
# Class for a single slot machine, where rewards are Gaussian
class GaussianBandit(object):
  def __init__(self, mean=0, stdev=1):
    self.mean = mean
    self.stdev = stdev

  def pull_lever(self):
    reward = np.random.normal(self.mean, self.stdev)
    return np.round(reward, 1)
```

```python
# Class to Simulate Game
class GaussianBanditGame(object):
  def __init__(self,bandits):
    self.bandits = bandits # An array containing the k - bandits
    np.random.shuffle(self.bandits) # Shuffled so that there is randomness
    self.reset_game()

  def reset_game(self):
    # To restartgame or to accomodate more bandits
    self.rewards = []
    self.total_reward = 0
    self.n_played = 0

  def play(self, choice):
    # To pull lever and calculate scores and append
    reward = self.bandits[choice - 1].pull_lever()
    self.rewards.append(reward)
    self.total_reward += reward
    self.n_played += 1
    return reward

  def user_play(self):
    self.reset_game()
    print("Game has Started. Enter 0 as an input to end the game.")
    while True:
      print(f"\n-- Round {self.n_played + 1}")
      choice = int(input(f"Choose a machine from 1 to {len(self.bandits)}: "))

      # To Quit
      if (choice < 1) or (choice > len(self.bandits)):
        print("Game has Ended!")
        break

      # To play
      else:
        reward = self.play(choice)
        print(f"Machine {choice} gave the reward: {reward}")
        avg_reward = self.total_reward/self.n_played
        print(f"Your Average Reward so far is: {avg_reward}")

    # Calculate final avergae
    if self.n_played != 0:
      print(f"Your total reward is {self.total_reward} after {self.n_played} rounds.\n With an average reward of {np.round(self.total_reward/self.n_played, 2)}")
```

```python
# Create some slot machine instances
slotA = GaussianBandit(5, 3)
slotB = GaussianBandit(7, 6)
slotC = GaussianBandit(4, 1)
```

```python
# Creating a game instance and simulating the game
game = GaussianBanditGame([slotA, slotB, slotC])
game.user_play()
```
